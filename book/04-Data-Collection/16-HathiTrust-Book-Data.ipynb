## Find a HathiTrust Volume ID

You can find a book's HathiTrust volume ID by following these steps:
- First, search the HathiTrust catalog. I recommend searching with the [Advanced Catalog Search](https://catalog.hathitrust.org/Search/Advanced).
- When you find the item that you're looking for, click "Limited (Search only)" or "Full View," depending on the copyright status of the item
- Then copy and paste the HathiID from the URL of the web page that you're taken to
    - For example:
        - https://babel.hathitrust.org/cgi/pt?id=mdp.39015029970129 -> `mdp.39015029970129`
        - https://babel.hathitrust.org/cgi/pt?id=uc2.ark:/13960/t7pn9353t&view=1up&seq=7 -> `uc2.ark:/13960/t7pn9353t`
    - When in doubt, look in the left sidebar for "Permanent link to this book." The volume ID will follow 2027
        - https://hdl.handle.net/2027/uc2.ark:/13960/t7pn9353t -> `uc2.ark:/13960/t7pn9353t`

## Make Word Frequency Pandas DataFrame from HathiTrust Volume(s)

````python
import pandas as pd
````

````python
from htrc_features import Volume
````

````python
volume_ids = ['mdp.39015029970129']
````

````python
all_tokens = []

for hathi_id in volume_ids:
    #Read in volume
    volume = Volume(hathi_id)
    
    #Make dataframe from token list -- do not include part of speech, sections, or case sensitivity
    token_df = volume.tokenlist(case=False, pos=False, drop_section=True)
    
    #Add book column
    token_df['book'] = volume.title
    
    #Add publication year column
    token_df['year'] = volume.year
    
    all_tokens.append(token_df)
````

````python
lost_df = pd.concat(all_tokens)
````

````python
lost_df
````

Change from multi-level index to regular index with `reset_index()`

````python
lost_df_flattened = lost_df.reset_index()
````

````python
lost_df_flattened 
````

````python
def add_chapter_titles(page):
    if page >= 11 and page < 35:
        return "01: The Girl Who Raised Pigeons"
    elif page >= 35 and page < 41:
        return "02: The First Day"
    elif page >= 41 and page < 63:
        return "03: The Night Rhonda Ferguson Was Killed"
    elif page >= 63 and page < 85:
        return "04: Young Lions"
    elif page >= 85 and page < 113:
        return "05: The Store"
    elif page >= 113 and page < 125:
        return "06: An Orange Line Train to Ballston"
    elif page >= 125 and page < 149:
        return "07: The Sunday Following Mother's Day"
    elif page >= 149 and page < 159:
        return "08: Lost in the City"
    elif page >= 159 and page < 184:
        return "09: His Mother's House"
    elif page >= 184 and page < 191:
        return "10: A Butterfly on F Street"
    elif page >= 191 and page < 209:
        return "11: Gospel"
    elif page >= 209 and page < 225:
        return "12: A New Man"
    elif page >= 225 and page < 237:
        return "13: A Dark Night"
    elif page >= 237:
        return "14: Marie"
````

````python
lost_df_flattened['chapter'] = lost_df_flattened['page'].apply(add_chapter_titles)
````

````python
lost_df_flattened['chapter'].value_counts()
````

Summarize token counts for each book

````python
lost_df_flattened.groupby(['chapter', 'lowercase'])[['count']].sum().reset_index()
````

````python
word_frequency_df = lost_df_flattened.groupby(['chapter', 'lowercase'])[['count']].sum().reset_index()
````

````python
word_frequency_df = word_frequency_df.drop(word_frequency_df[word_frequency_df['lowercase'].isin(STOPS)].index)
````

Remove punctuation

````python
word_frequency_df = word_frequency_df.drop(word_frequency_df[word_frequency_df['lowercase'].str.contains('[^A-Za-z\s]', regex=True)].index)
````

## TFIDF

````python
word_frequency_df = word_frequency_df.rename(columns={'lowercase': 'term','count': 'term_frequency'})
````

````python
word_frequency_df
````

### Calculate Document Frequency

In how many stories in the collection does this term appear?

````python
document_frequency_df = (word_frequency_df.groupby(['chapter','term']).size().unstack(fill_value=0) > 0).sum().reset_index()
````

````python
document_frequency_df = document_frequency_df.rename(columns={0:'document_frequency'})
````

````python
word_frequency_df = word_frequency_df.merge(document_frequency_df)
````

````python
word_frequency_df
````

As you can see in the DataFrame above, the term "abandoned" appears 2 times in the story "The Girl Who Raised Pigeons" (term frequency), and it appears in 3 different stories in the collection overall (document frequency).

### Calculate Total Number of Documents 

````python
total_number_of_documents = lost_df_flattened['chapter'].nunique()
````

### Calculate TFIDF

````python
import numpy as np
````

```
tfidf = term_frequency * inverse_document_frequency
```

```
inverse_document_frequency = log [ n / (df(t) + 1) ]
```

Scikitlearn version of [inverse document frequency with smoothing](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfTransformer.html#sklearn.feature_extraction.text.TfidfTransformer)

```
inverse_document_frequency = log [ (1 + n) / (1 + df(t)) ] + 1
```

````python
((numpy.log((1 + total_number_of_documents) / (1 + word_frequency_df['document_frequency']))) + 1)
````

````python
word_frequency_df['tfidf'] = word_frequency_df['term_frequency']  *       \
                            ((numpy.log((1 + total_number_of_documents) / (1 + word_frequency_df['document_frequency']))) + 1)
````

````python
#word_frequency_df['tfidf_standardized']= (word_frequency_df['tfidf'] - word_frequency_df['tfidf'].mean())/word_frequency_df['tfidf'].std()
````

````python
from sklearn import preprocessing 
````

````python
word_frequency_df['tfidf_sklearnnormalized'] = sklearn.preprocessing.normalize(word_frequency_df[['tfidf']], axis=0, norm='l2')
````

Find top 15 words with highest tfidf scores for each book

````python
word_frequency_df.sort_values(by=['chapter','tfidf_sklearnnormalized'], ascending=[True,False]).groupby(['chapter']).head(15)
````
